## Step 3 :

# Data Cleaning

```python
df_copy = df.copy()
```

```python
list(df_copy.columns.values)
```

```python
df_copy.rename(columns={'Goal Scored':'Goal_Scored',
 'Ball Possession %':'Ball_Possession',
 'Free Kicks':'Free_Kicks',
 'Pass Accuracy %':'Pass_Accuracy',
 'Distance Covered (Kms)':'Distance_Covered',
 'Fouls Committed':'Fouls_Committed',
 'Yellow Card':'Yellow_Card',
 'Yellow & Red':'Yellow & Red',
 'Man of the Match':'Man_of_the_Match',
 '1st Goal':'1st_Goals',
 'Goals in PSO':'Goals_in_PSO',
 'Own goals':'Own_goals',
 'Own goal Time':'Own_goal_Time'}, inplace=True)
```

```python
df_copy
```

## Cleaning 1
* Dealing with NaNs of 'Own_goals'

```python
df_copy.Own_goals=df_copy.Own_goals.fillna(0)
```

```python
df_copy
```

## Cleaning 2
* Convert 'Yes' and No' Values of 'Man_of_the_Match' and 'PSO' to numeric values

```python
df_copy['Man_of_the_Match']=df_copy['Man_of_the_Match'].map({'Yes':1,'No':0})
```

```python
df_copy.head()
```

```python
df_copy['PSO']=df_copy['PSO'].map({'Yes':1,'No':0})
```

## Cleaning 3
* Removing columns which won't be required for analysis

```python
columns=['Own_goal_Time','1st_Goals']
df_copy.drop(columns, axis=1, inplace=True)
```

```python
df_copy.info()
```

## Cleaning 4
* Data types mismatch fix

```python
df_copy.Own_goals=df_copy.Own_goals.astype('int64')
```

```python
df_copy.info()
```

```python
df_copy['Date'] = pd.to_datetime(df_copy['Date'])
```

```python
df_copy.info()
```

## Step 4 :

## Feature Engineering and Visualization

### Feature 1:
* Shot Accuracy

```python
df_copy['shot_accuracy']=df_copy['On-Target']/df_copy['Attempts']
```

```python
df_copy.groupby('Team')['Goal_Scored'].sum().plot(kind='bar',figsize=(12,8))
```

* Simple Graph Showing the most important feature while considering how well a team is while considering it's attacking prowess
* Belgium had scored the highest number of goals in the tournament.

### Feature 2:
* Goal Conversion Rate

```python
df_copy['coversion_rate']=df_copy['Goal_Scored']/df_copy['Attempts']
```

```python
df_copy.head()
```

```python
df_copy.groupby('Team')['coversion_rate'].mean().plot(kind='bar',figsize=(12,8))
plt.xlabel('Teams')
plt.ylabel('Conversion_Rate')
plt.title('Conversion')
plt.rcParams.update({'font.size': 15})
```

* An extension for previous result.
* Conversion Rate is a very important value to be considered while analyzing and deciding how good the attacking prowess of a team is
* France had the highest the Goal Conversion rate

### Feature 3:
* Additional features
* Created seperate dataframes for each feature for ease of plotting later on.

```python
df_goals=df_copy.groupby('Team')['Goal_Scored'].sum()
```

```python
df_goals1= pd.DataFrame(df_goals)
```

```python
matches_played=df_copy.groupby('Team')['Date'].count()
```

```python
df_mins1= pd.DataFrame(matches_played)
```

```python
df_mins1.rename(columns={'Date':'Matches_Played'},inplace=True)
```

```python
df_match2= pd.merge(df_goals1,df_mins1,how = 'outer', on = ['Team'])
```

```python
df_match2['Goals_per_Game']=df_match2['Goal_Scored']/df_match2['Matches_Played']
```

```python
oppg=df_copy.groupby('Opponent')['Goal_Scored'].sum()
```

```python
df_oppg= pd.DataFrame(oppg)
df_oppg
```

```python
df_oppg.index.names = ['Team']
df_oppg.rename(columns={'Goal_Scored':'Goals_Conceded'},inplace=True)
```

```python
df_match2= pd.merge(df_match2,df_oppg,how = 'outer', on = ['Team'])
```

```python
df_match2['Conceded_per_Game']=df_match2['Goals_Conceded']/df_match2['Matches_Played']
```

```python
df_match2
```

```python
x1=list(df_match2.index.values)
```

```python
df = pd.DataFrame({'Goals Scored': df_match2['Goals_per_Game'],'Goals Conceded':  df_match2['Conceded_per_Game']}, index=x1)
ax = df.plot.bar(figsize=(12,8))
plt.ylabel('Average Goals')
plt.xlabel('Teams')
plt.title('Attack v/s Defense')
```

* Decided to visually compare the attack and defense of the attack.
* Plotted Goals Scored per Game and Goals Conceded per game under the same index
* Panama had the most leaky defense and worst attack
* Belgium maintains the greatest Goals Scored and Goals Conceded ratio

## Feature 4
* Does teams's discipline affects the team's defense i.e does the fouls committed helps the team defensively or not.

```python
df_copy['Discipline']=df_copy['Yellow_Card']+df_copy['Red']+df_copy['Yellow & Red']
```

```python
df_disc=pd.DataFrame(df_copy.groupby('Opponent')['Discipline'].sum())
```

```python
df_disc.index.names = ['Team']
```

```python
df_match2= pd.merge(df_match2,df_disc,how = 'outer', on = ['Team'])
```

```python
df_match2.plot(kind='scatter', x='Discipline',y= 'Goals_Conceded',figsize=(8,8))
plt.xlabel('Discipline')
plt.ylabel('Goals_Conceded')
plt.title('Goal Conceded vs Discipline')
```

* Wanted to establish a relation between how disclipined a team is v/s How poor or good their defense is .
* Though I was not establish a definite trend between the two features but in general, a better disciplined defense means a better defense.

## Feature 5
* Attack v/s Possession

```python
df_copy.plot(kind='scatter', x='Ball_Possession',c='r',y= 'Attempts',figsize=(8,8), marker='o')
plt.xlabel('Ball Possession')
plt.ylabel('Attempts')
plt.title('Attempts vs Possession')
```

* Decided to plot this scatter graph to get a relation between Ball Possession and how good is the team moving forward
* Generally, higher the ball possession, greater the number of goal attempts.
* Correlation Coefficent of 0.54 is obtained.
* This indicates a strong positve correlation.
